```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("melb_data.csv")
```

```python
df
```

```python
x = df[["Rooms", "Bathroom", "Landsize", "Lattitude", "Longtitude"]]
y = df["Price"]
```

```python
x_train, x_validate, y_train, y_validate = train_test_split(x,y, random_state=0)
```

```python
x_train
```

```python
imputer = SimpleImputer()  #Este es el objeto cuya función será imputar el promedio de la columna a cada casilla q tenga NaN
x_train_imputed = pd.DataFrame(imputer.fit_transform(x_train)) #-->Agarra el x_train, el imputer lo rellena con los promedios en donde hay NaN y crea un nuevo df.
x_validate_imputed = pd.DataFrame(imputer.transform(x_validate)) #--> Ojo !! Acá no es idem anterior. Acá no se usa fit !!! Directamente transform.
```

```python
#IMPORTANTE!! El imputer SACA LOS ENCABEZADOS. Asique hay que volver a ponerselos:

x_train_imputed.columns = x_train.columns
x_validate_imputed.columns = x_validate.columns
```

```python
x_train_imputed
```

```python
x_validate_imputed
```

```python
def Error (x_train , x_validate, y_train, y_validate, n_estimators):
    
    modelo = RandomForestRegressor(n_estimators=n_estimators, random_state=0)
    modelo.fit(x_train, y_train)
    predicciones = modelo.predict(x_validate)
    MAE = mean_absolute_error(y_validate, predicciones)
    return MAE
```

```python
Min_MAE = 1000000000

for n in range [1100000]:
      
    new_MAE = Error(x_train_imputed, x_validate_imputed, y_train, y_validate, n)
    
    if new_MAE < min_MAE:
        
        n_estimators = n
        Min_MAE = new_MAE
        
print("Optimo n_estimators")
print(n_estimators)
print("Minimo error en este modelo")
print(Min_MAE)
```
