```python
import pandas as pd

df = pd.read_csv("melb_data.csv")
```

```python
df.dropna(axis=0)
```

```python
x = df[["Rooms", "Bathroom", "Landsize", "Lattitude", "Longtitude"]]
```

```python
x
```

```python
y = df[["Price"]]
y
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
x_train , x_validate , y_train , y_validate = train_test_split(x,y) #Creo la data para ajustar el modelo y para validarlo. 

#Variables (Data) CREADAS.
```

```python
def ObtenerErrorValidate(max_leaf_nodes, x_train, x_validate, y_train, y_validate):
    
    modelo = DecisionTreeRegressor(max_leaf_nodes= max_leaf_nodes, random_state=0) #Creo el arbol, con un máx de divisiones y el random state para obtener siempre los mismos nros (doy repetitibilidad)
    modelo.fit(x_train,y_train) #Ajuste del modelo con la Train Data.
    predicciones = modelo.predict(x_validate)
    MAE = mean_absolute_error(y_validate , predicciones)
    return MAE

def ObtenerErrorTrain(max_leaf_nodes, x_train, y_train):
    
    modelo = DecisionTreeRegressor(max_leaf_nodes= max_leaf_nodes, random_state=0) #Creo el arbol, con un máx de divisiones y el random state para obtener siempre los mismos nros (doy repetitibilidad)
    modelo.fit(x_train,y_train) #Ajuste del modelo con la Train Data.
    predicciones = modelo.predict(x_train)
    MAE = mean_absolute_error(y_train , predicciones)
    return MAE
```

```python
ObtenerErrorValidate(10000, x_train, x_validate, y_train, y_validate)
```

```python
ObtenerErrorTrain(10000,x_train,y_train) 
```

```python
for max_leaf_nodes in [5, 100, 1000, 10000, 15000, 20000]:
   
    ErrorValidate = ObtenerErrorValidate(max_leaf_nodes, x_train, x_validate, y_train, y_validate)
    ErrorTrain = ObtenerErrorTrain(max_leaf_nodes, x_train, y_train)
    
    print("Nodos máximos aceptados: %d \t MAE in ValidateData: %d \t MAE in TrainingData: %d" %(max_leaf_nodes, ErrorValidate, ErrorTrain))
```
