```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
#ADEMÁS DE REEMPLAZAR POR EL PROMEDIO, LO QUE SE HACE ES AGREGAR UNA COLUMNA QUE AVISA SI EL VALOR FALTABA O NO.
```

```python
df = pd.read_csv("melb_data.csv")
df
```

```python
x = df[["Rooms", "Bathroom", "Landsize", "Lattitude", "Longtitude"]]
y = df["Price"]
```

```python
x
```

```python
y
```

```python
x_train, x_validate, y_train, y_validate = train_test_split(x, y, random_state=0)
```

```python
#Creamos una copia, para no modificar la data original 
x_train_plus = x_train.copy()
x_validate_plus = x_validate.copy()
```

```python
# ¿En que columnas faltan valores ? 

columnas_train_na = []
columnas_validate_na = []

for col in x_train_plus:
    if x_train_plus[col].isnull().any():
        
        columnas_train_na = columnas_train_na + [col]
        

for col in x_validate_plus:
    if x_validate_plus[col].isnull().any():
        
        columnas_validate_na = columnas_validate_na + [col]
        
        
print(columnas_train_na)
print(columnas_validate_na)
```

```python
#Ahora lo que vamos a hacer es ver donde hay NaN y crear una nueva columna donde diga si el NaN estaba o no. 

#SUPER MEGA IMPORTANTE. OJO

#Puede pasar, que cuando se haga la division de data con 'train_test_split', una mitad de la data quede con NaN y otra mitad 
# de la data no. Ahí tenemos un problema, porque puede crear por ejemplo para la x_train una nueva columna que se llame 
# 'Rooms_missing_values', pero para la x_validate eso no sucederá porque en la x_validate no falta ningun número en la columna
# 'Rooms'. Es decir, solo x_train tiene NaN en la columna Rooms.

#¿Qué sucede entonces? --> Se esta ajustando (fit) un modelo con 6 variables x Y luego se evalúa en 5.     
```

```python
#Entonces: VAMOS A CREAR ESA COLUMNA 'ROOMS_MISSING_VALUES' para ambas partes de la data, tanto la train como la validate.

# ¿Cómo? : Así

for col in columnas_train_na:  #columnas_train_na = ["Rooms"]
    
    x_train_plus[col + "_missing_value"] = x_train_plus[col].isnull()  #  "x_train_plus.isnull()" devuelve un Booleano True o False
    x_validate_plus[col + "_missing_value"] = x_validate_plus[col].isnull()
    
# En la ultima linea de código, va a crear la columna "Rooms_missing_value" para validate. 
# Podría pasar que la validate data no tenga ningun NaN en Rooms ! (En ese caso 'Rooms_missing_values' se llena de Falses)
# Pero es fundamental que la x_train y la x_validate TENGAN LA MISMA CANTIDAD DE COLUMNAS.
```

```python
# AHORA SI, una vez que tengo la columna extra creada (la cual avisa si existia un NaN o no) puedo imputar.
# Es decir, reemplazar el NaN por el promedio de la columna. 
```

```python
#Armo el imputer para que ÉL OBJETO SOLITO ponga el promedio en lugar de los NaN

imputer = SimpleImputer()
x_train_plus_imputed = pd.DataFrame(imputer.fit_transform(x_train_plus))
x_validate_plus_imputed = pd.DataFrame(imputer.fit_transform(x_validate_plus))
        
```

```python
x_train_plus_imputed.columns = x_train_plus.columns
x_validate_plus_imputed.columns = x_validate_plus.columns
```

```python
x_train_plus_imputed
```

```python
x_validate_plus_imputed
```

```python
def Error (x_train , x_validate, y_train, y_validate):
    
    modelo = RandomForestRegressor(random_state=0)
    modelo.fit(x_train, y_train)
    predicciones = modelo.predict(x_validate)
    MAE = mean_absolute_error(y_validate, predicciones)
    return MAE
```

```python
print(Error(x_train_plus_imputed, x_validate_plus_imputed, y_train, y_validate))
```
