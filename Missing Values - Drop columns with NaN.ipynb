```python
#CLAVE!!! SI ALGUNA COLUMNA TIENE UN NaN --> El sig. código hace una lista con las columnas que tienen algun NaN

columns_na = []

for col in df2.columns:     #Para las columnas del df2:
    
    if df2[col].isnull().any():  #Si la columna de df2 tiene algun/cualquier valor nulo (any is null)
        
        columns_na = columns_na + [col]  #FUNDAMENTAL acá poner [col] entre corchetes. Porque es un elemento más de la lista
```

```python
def Error (x_train , x_validate, y_train, y_validate):
    
    modelo = RandomForestRegressor(random_state=0)
    modelo.fit(x_train, y_train)
    predicciones = modelo.predict(x_validate)
    MAE = mean_absolute_error(y_validate, predicciones)
    return MAE
```
